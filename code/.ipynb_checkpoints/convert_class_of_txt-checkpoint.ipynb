```python
"""

this selects files that only have these class numbers

"""


chosen_class = ['5', '1', '11', '17', '2', '0', '22', '24']
source_dir = './valid/'
target_dir = './new_valid/'

files = os.listdir(source_dir)

for file in files:
    if file.split('.')[-1] == 'txt':
        f = open(source_dir + file, 'r')
        #cclass = f.read(1)
        line = f.readline()
    # create class var
        
        cclass = line.split(' ')[0]
        
        
        if cclass in chosen_class:
            # copy txt file 
            
            print(file)
            copyfile(source_dir + file, target_dir + file)
            
            file_root = ['.'.join(file.split('.')[:-1]), 'jpg']
            jpg_file = '.'.join(file_root)
            print(jpg_file)
            copyfile(source_dir + jpg_file, target_dir + jpg_file)
            
        #line = f.readline()
        f.close()
```
